Downsampling in loop and modifying it

```python
%cd /content/bengali_female_english/bengali_female_english/bengali_femaleenglish/

new_directory = '/content/bengali_female/wav/'
path = 'wav'
import librosa
import soundfile as sf
import os
directory = os.fsencode(path)
    
for file in os.listdir(directory):
     filename = os.fsdecode(file)
     if filename.endswith(".wav") : 
        new_path = new_directory+filename
        original_path = '/content/bengali_female_english/bengali_female_english/bengali_femaleenglish/wav/'+filename
        y, sr = librosa.load( original_path, sr=16000 )
        sf.write( new_path, y, 16000)
        #print(os.path.join(directory, filename))
     else: 0
```

```python
%cd /content/bengali_male_english/bengali_male_english/

new_directory = '/content/bengali_male/wav/'
path = 'wav'
import librosa
import soundfile as sf
import os
directory = os.fsencode(path)
    
for file in os.listdir(directory):
     filename = os.fsdecode(file)
     if filename.endswith(".wav") : 
        new_path = new_directory+filename
        original_path = '/content/bengali_male_english/bengali_male_english/wav/'+filename
        y, sr = librosa.load( original_path, sr=16000 )
        sf.write( new_path, y, 16000)
        #print(os.path.join(directory, filename))
     else: 0
```

Training part, --train_cudnn True can be omitted, if not using gpu


```python
%cd /content/DeepSpeech
!python3 DeepSpeech.py --train_cudnn True --early_stop True --es_epochs 6 --n_hidden 2048 --epochs 5 \
  --export_dir /content/drive/MyDrive/models/ --checkpoint_dir /content/drive/MyDrive/model_checkpoints/deepspeech-0.9.3-checkpoint/ \
  --train_files /content/train.csv --dev_files /content/dev.csv --test_files /content/test.csv \
  --learning_rate 0.0001 --train_batch_size 32 --test_batch_size 32 --dev_batch_size 32 --export_file_name 'ft_bengali_female_model' \
  --augment reverb[p=0.2,delay=50.0~30.0,decay=10.0:2.0~1.0] \
  --augment volume[p=0.2,dbfs=-10:-40] \
  --augment pitch[p=0.2,pitch=1~0.2] \
  --augment tempo[p=0.2,factor=1~0.5]
```
